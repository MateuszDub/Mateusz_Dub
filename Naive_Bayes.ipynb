# Import libraries

```
import numpy as np
import pandas as pd
import sklearn

np.random.seed(42)
np.set_printoptions(precision=6, suppress=True)
```

# Generating data

```
weather = ['sunny', 'rainy', 'cloudy', 'rainy', 'sunny', 'sunny', 'cloudy',
           'cloudy','sunny']
temperature= ['hot', 'cold', 'hot', 'hot', 'hot', 'warm', 'warm','hot','cold']
stroll = ['yes', 'no', 'yes', 'no', 'yes', 'yes', 'no', 'yes', 'no']

raw_df = pd.DataFrame(data={'weather': weather, 'temperature': temperature,
                            'stroll':stroll})
df = raw_df.copy()
df
```

# Preprocesing data

```
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df['stroll'] = encoder.fit_transform(stroll)
df
```

```
df = pd.get_dummies(df, columns=['weather', 'temperature'], drop_first=True)
df 
```

```
data = df.copy()
target = data.pop('stroll')
data
```

```
target
```

# Naive Bayes classifier

```
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(data, target)
```

```
model.score(data, target)
```
